```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv('Pokemon.csv')
pokemon.shape
```

```python
pokemon.head()
```

```python
g = nx.Graph()

# 앞에서 100개만 먼저 불러와 봅니다.
g = nx.from_pandas_edgelist(pokemon[:100], source='Name', target='Type 1')
G = nx.from_pandas_edgelist(pokemon, source='Name', target='Type 1')
print(nx.info(g))
```

```python
nx.draw(g, with_labels=True)
plt.show()
```

```python
plt.figure(figsize=(20, 20))
# pos = nx.spring_layout(g, k=0.15)  # position 생성
pos = nx.spring_layout(G, k=0.15)
# nx.draw_networkx(g,pos,node_size=25, node_color='red')
nx.draw_networkx(G,pos,node_size=25, node_color='red')
plt.show()
```

#### type 2 그리기

```python
g = nx.Graph()
g = nx.from_pandas_edgelist(pokemon[:100], source='Name', target='Type 2')
print(nx.info(g))
```

```python
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(g, k=0.3)  # k: 뭉쳐있고, 퍼져있는 비율을 정해줌.
nx.draw_networkx(g,pos,node_size=25, node_color='red')
plt.show()
```
